```python
import pandas as pd
import os
```

```python
new_globug_dataset_path = "/home/varumuga/scratch/Thesis/Replications/Globug/globug_bl_dataset/new_dataset"
```

```python
if not os.path.isdir(new_globug_dataset_path):
    os.makedirs(new_globug_dataset_path)
```

```python
train_test_split_bl_dataset_path = "/home/varumuga/scratch/Thesis/bench_bl_dataset/bl_dataset"
```

```python
data_types = ["cross_project", "within_project", "cross_project_and_within_project"]
```

```python
projects = ['AspectJ', 'Eclipse_Platform_UI', 'JDT', 'SWT', 'Birt', 'Tomcat']
```

```python
combined_df = pd.read_pickle("/home/varumuga/scratch/Thesis/Replications/Globug/globug_bl_dataset/combined_dataset/allBugReportsProcessed.pickle")
```

```python
combined_df = combined_df.set_index(['bug_id', 'commit', 'project_name']) 
```

```python
for data_type in data_types:
    train_dfs = []
    test_dfs = []
    for project_folder in projects:
        dataset_split = os.path.join(train_test_split_bl_dataset_path, data_type, project_folder)
        train_br_df = pd.read_csv(os.path.join(dataset_split, f"{project_folder}_train_br.csv"), sep='\t')
        
        test_br_df = pd.read_csv(os.path.join(dataset_split, f"{project_folder}_test_br.csv"), sep='\t')
        
        train_br_df = train_br_df.set_index(['bug_id', 'commit', 'project_name'])
        
        test_br_df = test_br_df.set_index(['bug_id', 'commit', 'project_name'])
        
        train_br_dfs_processed = combined_df[combined_df.index.isin(train_br_df.index)]
        
        test_br_dfs_processed = combined_df[combined_df.index.isin(test_br_df.index)]
        
        print(f"Train old size {train_br_df.shape} New size {train_br_dfs_processed.shape}")
        
        print(f"Test old size {test_br_df.shape} New size {test_br_dfs_processed.shape}")
        
        new_dt_folder = os.path.join(new_globug_dataset_path, data_type, project_folder)
        if not os.path.isdir(new_dt_folder):
            os.makedirs(new_dt_folder)
        
        train_br_dfs_processed.to_csv(os.path.join(new_dt_folder, f"{project_folder}_train_br.csv"), sep='\t')
        train_br_dfs_processed.to_pickle(os.path.join(new_dt_folder, f"{project_folder}_train_br.pickle"))
        
        test_br_dfs_processed.to_csv(os.path.join(new_dt_folder, f"{project_folder}_test_br.csv"), sep='\t')
        test_br_dfs_processed.to_pickle(os.path.join(new_dt_folder, f"{project_folder}_test_br.pickle"))
```

```python
combined_sc = pd.read_parquet("/home/varumuga/scratch/Thesis/Replications/Globug/globug_bl_dataset/combined_dataset/allSourceCodesProcessed")
```

```python
combined_sc_ = combined_sc.set_index(["cid", "file_content"])
```

```python
combined_sc_unprocessed = pd.read_csv("/home/varumuga/scratch/Thesis/Replications/Globug/globug_bl_dataset/combined_dataset/allSourceCodes.csv")
```

```python
combined_sc_unprocessed_ = combined_sc_unprocessed.set_index(["cid", "file_content"])
```

```python
combined_sc_unprocessed_.shape, combined_sc_.shape
```

```python
combined_sc_ = combined_sc_[~combined_sc_.index.duplicated(keep='first')]
```

```python
combined_sc_.shape, combined_sc_unprocessed_.shape
```

```python
combined_sc = pd.concat([combined_sc_, combined_sc_unprocessed_], axis=1)
```

```python
combined_sc.head()
```

```python
combined_sc = combined_sc.reset_index(level=['cid', 'file_content'])
combined_sc = combined_sc.set_index(['cid', 'project_name'])
```

```python
combined_sc.head()
```

```python
features_col_to_drop = ['Unnamed: 0', 'bug_recency', 'rVSM_similarity',
       'bug_frequency', 'classname_similarity', 'collab_filter']
```

```python
for data_type in data_types:
    train_dfs = []
    test_dfs = []
    for project_folder in projects:
        dataset_split = os.path.join(train_test_split_bl_dataset_path, data_type, project_folder)
        train_features_df = pd.read_csv(os.path.join(dataset_split, f"{project_folder}_train_features.csv"))
        
        test_features_df = pd.read_csv(os.path.join(dataset_split, f"{project_folder}_test_features.csv"))
        
        train_features_df = train_features_df.set_index(["cid", "project_name"])
        
        test_features_df = test_features_df.set_index(["cid", "project_name"])
        
        train_features_df = train_features_df.drop(columns=features_col_to_drop)
        
        test_features_df = test_features_df.drop(columns=features_col_to_drop)
        
        train_features_dfs_processed = pd.concat([combined_sc[combined_sc.index.isin(train_features_df.index)], train_features_df], axis=1) 
        
        test_features_dfs_processed = pd.concat([combined_sc[combined_sc.index.isin(test_features_df.index)], test_features_df], axis=1) 
        
        train_features_dfs_processed = train_features_dfs_processed.reset_index(level=['cid', 'project_name'])

        test_features_dfs_processed = test_features_dfs_processed.reset_index(level=['cid', 'project_name'])
        
        print(f"Train old size {train_features_df.shape} New size {train_features_dfs_processed.shape}")
        
        print(f"Test old size {test_features_df.shape} New size {test_features_dfs_processed.shape}")
        
        new_dt_folder = os.path.join(new_globug_dataset_path, data_type, project_folder)
        if not os.path.isdir(new_dt_folder):
            os.makedirs(new_dt_folder)
        
        # print(os.path.join(new_dt_folder, f"{project_folder}_train_sc"))
        # print(train_features_dfs_processed.columns)
        
        print(f"Writing train df to {os.path.join(new_dt_folder, f'{project_folder}_train_sc')}")
        
        try:
            train_features_dfs_processed.to_parquet(os.path.join(new_dt_folder, f"{project_folder}_train_sc"), row_group_offsets=20000000)
        except:
            train_features_dfs_processed.to_csv(os.path.join(new_dt_folder, f"{project_folder}_train_sc.csv"))
        # train_br_dfs_processed.to_pickle(os.path.join(new_dt_folder, f"{project_folder}_train_br.pickle"))
        
         # print(f"Writing test df to {os.path.join(new_dt_folder, f'{project_folder}_test_sc')}")
        try:
            test_features_dfs_processed.to_parquet(os.path.join(new_dt_folder, f"{project_folder}_test_sc"), row_group_offsets=20000000 )
        except:
            test_features_dfs_processed.to_csv(os.path.join(new_dt_folder, f"{project_folder}_test_sc.csv"))
        # test_br_dfs_processed.to_pickle(os.path.join(new_dt_folder, f"{project_folder}_test_br.pickle"))
        # break
```
